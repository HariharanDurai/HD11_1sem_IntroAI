String Operations

length of string , particular character in a string, slicing a string, Negative indexing, Upper case , lower case, reverse

```python
Hello = "Harry is in SRM university"
print(len(Hello))
print(Hello.find('R'))
print (Hello.upper())
print(Hello.lower())
print(Hello[::-1]) # This is how you reverse a string using slicing
print("".join(reversed(Hello)))
print(reversed(Hello))
```

```python
Hello = "Harry is in SRM university"

# Get the first 5 characters
print(Hello[0:5])

# Get characters from index 6 to the end
print(Hello[6:])

# Get characters from the beginning to index 10
print(Hello[:10])

# Get characters with a step of 2
print(Hello[::2])

# Reverse the string using slicing (as we saw before)
print(Hello[::-1])

# Get characters from index 6 to 15 with a step of 3
print(Hello[6:15:3])
```
